PHẦN 1: Thống kê dữ liệu


```python
#1. Hãy sắp xếp dữ liệu điểm DH1 theo thứ tự tăng dần
import pandas as pd

# Đọc dữ liệu từ tệp CSV
df = pd.read_csv('processed_dulieuxettuyendaihoc.csv')

# Sắp xếp dữ liệu theo cột DH1 tăng dần
df_sorted = df.sort_values(by='DH1', ascending=True)

# Hiển thị 10 dòng đầu tiên sau khi sắp xếp
df_sorted

```

```python
#2. Hãy sắp xếp dữ liệu điểm DH2 tăng dần theo nhóm giới tính

# Sắp xếp dữ liệu theo giới tính (GT) trước, sau đó sắp xếp điểm DH2 tăng dần trong mỗi nhóm
df_sorted = df.sort_values(by=['GT', 'DH2'], ascending=[True, True])

# Hiển thị một vài dòng đầu của kết quả đã sắp xếp
df_sorted
```

```python
#3. Hãy tạo pivot-table để thống kê các giá trị count, sum, mean, median, min, max, std, Q1, Q2 và Q3 của DH1 theo KT

# Định nghĩa các hàm tính Q1 và Q3
def q1(x):
    return x.quantile(0.25)

def q3(x):
    return x.quantile(0.75)

# Tạo pivot table thống kê DH1 theo KT
pivot_table = df.groupby('KT')['DH1'].agg(
    count='count',
    sum='sum',
    mean='mean',
    median='median',
    min='min',
    max='max',
    std='std',
    Q1=q1,
    Q2='median',
    Q3=q3
)

# Hiển thị kết quả
pivot_table
```

```python
#4. Hãy tạo pivot-table để thống kê các giá trị count, sum, mean, median, min, max, std, Q1,
#Q2 và Q3 của DH1 theo KT và KV

# Định nghĩa các hàm để tính phân vị (Q1, Q2, Q3)
def q1(x):
    return x.quantile(0.25)

def q2(x):
    return x.quantile(0.50)

def q3(x):
    return x.quantile(0.75)

# Tạo bảng pivot-table thống kê theo KT
pivot_table = df.groupby('KT')['DH1'].agg(['count', 'sum', 'mean', 'median', 'min', 'max', 'std', q1, q2, q3])

# Đổi tên các cột cho rõ ràng
pivot_table.columns = ['count', 'sum', 'mean', 'median', 'min', 'max', 'std', 'Q1', 'Q2', 'Q3']

# Xuất kết quả ra file CSV
pivot_table.to_csv('pivot_table_DH1_by_KT.csv')
pivot_table
```

PHẦN 3: Trực quan hóa dữ liệu theo nhóm phân loại

```python
# 1. Trực quan dữ liệu học sinh nữ trên các nhóm XL1, XL2, XL3 dạng unstacked
# Gợi ý
# - Lọc dữ liệu giới tính là nữ
# - Oy: Chiều cao biểu đồ cột thể hiển số lượng học sinh theo xếp loại
# - Màu sắc thể hiện giá trị xếp loại: [Y, TB, K, G, XS]
# - Ox: thể hiện nhóm XL1, XL2 và XL3
import matplotlib.pyplot as plt
import seaborn as sns


# 2. Lọc dữ liệu học sinh nữ (GT == 'F')
df_female = df[df['GT'] == 'F']

# 3. Chuyển đổi dữ liệu sang dạng long-format để dễ vẽ biểu đồ
# Biến XL1, XL2, XL3 thành các giá trị trong một cột 'Nhom'
df_melted = df_female.melt(id_vars=['STT'], value_vars=['XL1', 'XL2', 'XL3'],
                           var_name='Nhom', value_name='XepLoai')

# 4. Tính toán số lượng học sinh theo từng Nhóm và Xếp loại
df_counts = df_melted.groupby(['Nhom', 'XepLoai']).size().reset_index(name='SoLuong')

# Định nghĩa thứ tự xếp loại
rating_order = ['Y', 'TB', 'K', 'G', 'XS']
df_counts['XepLoai'] = pd.Categorical(df_counts['XepLoai'], categories=rating_order, ordered=True)

# 5. Vẽ biểu đồ cột unstacked (grouped bar chart)
plt.figure(figsize=(10, 6))
sns.barplot(data=df_counts, x='Nhom', y='SoLuong', hue='XepLoai', hue_order=rating_order)

# Thêm tiêu đề và nhãn
plt.title('Số lượng học sinh nữ theo xếp loại trong các nhóm XL1, XL2, XL3', fontsize=14)
plt.xlabel('Nhóm (XL1, XL2, XL3)', fontsize=12)
plt.ylabel('Số lượng học sinh', fontsize=12)
plt.legend(title='Xếp loại', loc='upper right')
plt.grid(axis='y', linestyle='--', alpha=0.7)


plt.show()
```

```python
#2. Trực quan dữ liệu KQXT trên nhóm học sinh có khối thi A, A1, B thuộc khu vực 1, 2

# 2. Xử lý và Lọc dữ liệu
# Chuyển đổi KV sang string để đảm bảo tính nhất quán khi lọc
df['KV'] = df['KV'].astype(str)

# Thiết lập điều kiện lọc
condition_kt = df['KT'].isin(['A', 'A1', 'B'])
condition_kv = df['KV'].isin(['1', '2'])

# Lấy dữ liệu thỏa mãn
filtered_df = df[condition_kt & condition_kv]

# 3. Trực quan hóa
if not filtered_df.empty:
    plt.figure(figsize=(8, 6))

    # Thống kê số lượng Đậu/Rớt
    # Giả sử 0 là Rớt, 1 là Đậu
    kq_counts = filtered_df['KQXT'].value_counts().sort_index()

    # Vẽ biểu đồ cột
    colors = ['#F44336', '#4CAF50'] # Đỏ cho Rớt, Xanh cho Đậu
    ax = kq_counts.plot(kind='bar', color=colors, edgecolor='black')

    plt.title('Phân bố KQXT của học sinh Khối A, A1, B - KV 1, 2')
    plt.xlabel('Kết quả (0: Rớt, 1: Đậu)')
    plt.ylabel('Số lượng thí sinh')
    plt.xticks(rotation=0)

    # Hiển thị con số cụ thể trên từng cột
    for i, v in enumerate(kq_counts):
        ax.text(i, v + 0.5, str(v), ha='center', va='bottom', fontweight='bold')

    plt.grid(axis='y', linestyle='--', alpha=0.5)
    plt.show()
else:
    print("Không tìm thấy dữ liệu phù hợp với điều kiện lọc.")
```

```python
#4. Trực quan dữ liệu số lượng thí sinh đậu, rớt trên từng nhóm khối thi

# 2. Tạo cột nhãn 'KQXT_Label' để hiển thị 'Đậu'/'Rớt' thay vì 1/0
# Giả sử: 1.0 là Đậu, 0.0 là Rớt
df['KQXT_Label'] = df['KQXT'].map({1.0: 'Đậu', 0.0: 'Rớt'})

# 3. Thiết lập kích thước biểu đồ
plt.figure(figsize=(10, 6))

# 4. Vẽ biểu đồ cột (Countplot)
# Trục x là Khối thi (KT), màu sắc (hue) phân biệt theo Kết quả (KQXT_Label)
ax = sns.countplot(data=df, x='KT', hue='KQXT_Label', palette='viridis')

# 5. Thêm tiêu đề và nhãn cho các trục
plt.title('Số lượng thí sinh Đậu/Rớt theo từng Khối thi', fontsize=15)
plt.xlabel('Khối thi', fontsize=12)
plt.ylabel('Số lượng thí sinh', fontsize=12)
plt.legend(title='Kết quả')

# 6. Hiển thị số lượng cụ thể trên đầu mỗi cột
for p in ax.patches:
    # Lấy chiều cao của cột (số lượng)
    height = p.get_height()
    # Kiểm tra nếu height là số (để tránh lỗi với dữ liệu trống)
    if height > 0:
        ax.annotate(f'{int(height)}',
                    (p.get_x() + p.get_width() / 2., height),
                    ha = 'center', va = 'center',
                    xytext = (0, 5),
                    textcoords = 'offset points')

# 7. Hiển thị biểu đồ
plt.show()
```

```python
#Hãy tạo biến phân loại (phanlopt1) cho môn toán (T1) như sau:
#a. Từ 0 đến dưới 5 = kém (ký hiệu “k”)
#b. Từ 5 đến dưới 7 = trung bình (ký hiệu “tb”)
#c. Từ 7 đến dưới 8 = khá (ký hiệu “k”)
#d. Từ 8 trở lên = giỏi (ký hiệu “g”)

#  Định nghĩa hàm phân loại
def phan_loai(t1):
    if t1 < 5:
        return 'k'   # Kém
    elif t1 < 7:
        return 'tb'  # Trung bình
    elif t1 < 8:
        return 'k'   # Khá
    else:
        return 'g'   # Giỏi

# Áp dụng hàm vào cột T1 để tạo biến mới phanlopt1
df['phanlopt1'] = df['T1'].apply(phan_loai)

# Hiển thị 5 dòng đầu để kiểm tra
print(df[['T1', 'phanlopt1']].head())
```

```python

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

# --- Bước 1: Đọc dữ liệu ---
t1 = df['T1']

# --- Bước 2: Tính toán các đại lượng thống kê ---
# Các đại lượng cơ bản
mean_val = t1.mean()       # Trung bình
median_val = t1.median()   # Trung vị
mode_val = t1.mode()[0]    # Yếu vị (xuất hiện nhiều nhất)
min_val = t1.min()         # Nhỏ nhất
max_val = t1.max()         # Lớn nhất
std_val = t1.std()         # Độ lệch chuẩn

# Các đại lượng cho Box-Plot
q1 = t1.quantile(0.25)     # Tứ phân vị thứ 1 (25%)
q3 = t1.quantile(0.75)     # Tứ phân vị thứ 3 (75%)
iqr = q3 - q1              # Khoảng tứ phân vị (Interquartile Range)

# Tính Skewness và Kurtosis để đánh giá hình dáng phân phối
skew_val = t1.skew()       # Độ lệch
kurt_val = t1.kurt()       # Độ nhọn

# --- Bước 3: In kết quả kèm chú thích ---
print("=== KẾT QUẢ KHẢO SÁT PHÂN PHỐI BIẾN T1 ===")
print(f"1.  Mean (Trung bình) = {mean_val:.3f} -> Điểm trung bình môn Toán là khoảng 5.95.")
print(f"2.  Median (Trung vị) = {median_val:.3f} -> 50% sinh viên có điểm dưới 5.85 và 50% trên 5.85.")
print(f"3.  Mode (Yếu vị)     = {mode_val:.3f} -> Mức điểm 5.2 xuất hiện nhiều nhất.")
print(f"4.  Min (Thấp nhất)   = {min_val} -> Sinh viên thấp điểm nhất là 2.4.")
print(f"5.  Max (Cao nhất)    = {max_val} -> Sinh viên cao điểm nhất là 9.3.")
print(f"6.  Std Dev           = {std_val:.3f} -> Điểm số dao động trung bình khoảng 1.6 điểm quanh mức trung bình.")
print(f"7.  Skewness          = {skew_val:.3f} -> Âm nhẹ (< 0), phân phối hơi lệch trái (đuôi bên trái dài hơn).")
print(f"8.  Kurtosis          = {kurt_val:.3f} -> Âm (< 0), phân phối đỉnh bẹt (thấp hơn phân phối chuẩn).")
print("-" * 30)
print("=== CÁC ĐẠI LƯỢNG TRONG BOX-PLOT ===")
print(f"9.  Q1 (25%)          = {q1}")
print(f"10. Q3 (75%)          = {q3}")
print(f"11. IQR               = {iqr:.3f}")

# --- Bước 4: Vẽ biểu đồ ---
plt.figure(figsize=(15, 5))

# 4.1 Biểu đồ Box-Plot
plt.subplot(1, 3, 1)
plt.boxplot(t1, vert=True, patch_artist=True)
plt.title('Box-Plot: Xác định ngoại lai & phân tán')
plt.grid(True, linestyle='--', alpha=0.5)

# 4.2 Biểu đồ Histogram
plt.subplot(1, 3, 2)
sns.histplot(t1, kde=True, bins=10, color='skyblue')
plt.title('Histogram: Hình dáng phân phối')

# 4.3 Biểu đồ QQ-Plot
plt.subplot(1, 3, 3)
stats.probplot(t1, dist="norm", plot=plt)
plt.title('QQ-Plot: Kiểm tra phân phối chuẩn')

plt.tight_layout()
plt.show()
```

```python
#Hãy mô tả và khảo sát phân phối cho biến T1 trên từng nhóm phân lớp (phanlopT1)
#Gợi ý
#- Trực quan hóa biểu đồ Box-plot, histogram và QQ-plot theo phân nhóm là giá trị
#của ‘phanlopT1’.

# 1. Đọc dữ liệu và tạo biến phanlopT1
def phan_loai(t1):
    if t1 < 5: return 'k'
    elif t1 < 7: return 'tb'
    elif t1 < 8: return 'k'
    else: return 'g'
df['phanlopT1'] = df['T1'].apply(phan_loai)

# 2. Thống kê mô tả
print(df.groupby('phanlopT1')['T1'].describe())

# 3. Vẽ Box-plot so sánh các nhóm
plt.figure(figsize=(10, 6))
sns.boxplot(x='phanlopT1', y='T1', data=df, palette="Set2")
plt.title('So sánh phân phối T1 qua Box-plot')
plt.show()

# 4. Vẽ Histogram và QQ-plot cho từng nhóm
groups = df['phanlopT1'].unique()
fig, axes = plt.subplots(len(groups), 2, figsize=(12, 4 * len(groups)))

for i, group in enumerate(groups):
    subset = df[df['phanlopT1'] == group]['T1']
    # Histogram
    sns.histplot(subset, kde=True, ax=axes[i, 0], color='skyblue')
    axes[i, 0].set_title(f'Histogram nhóm {group}')
    # QQ-plot
    stats.probplot(subset, dist="norm", plot=axes[i, 1])
    axes[i, 1].set_title(f'QQ-plot nhóm {group}')

plt.tight_layout()
plt.show()
```

```python

 #Hãy khảo sát tương quan giữa biến DH1 theo biến T1
#Gợi ý
#- Nhận xét giá trị Covariance hoặc Correlation
#- Vẽ biểu đồ Scatter thể hiện liên hệ của biến phụ thuộc DH1 theo biến độc lập T1


# 1. Đọc dữ liệu
# 2. Tính toán các chỉ số tương quan
# Covariance
covariance = df['T1'].cov(df['DH1'])
# Correlation
correlation = df['T1'].corr(df['DH1'])

print(f"Covariance (Hiệp phương sai): {covariance:.3f}")
print(f"Correlation (Hệ số tương quan): {correlation:.3f}")

# 3. Vẽ biểu đồ Scatter kèm đường xu hướng (Regression Line)
plt.figure(figsize=(10, 6))

# scatter_kws: chỉnh độ trong suốt của điểm
# line_kws: chỉnh màu đường xu hướng
sns.regplot(x='T1', y='DH1', data=df,
            line_kws={'color': 'red'},
            scatter_kws={'alpha': 0.6, 'color': 'blue'})

plt.title(f'Biểu đồ Scatter: T1 vs DH1 (Correlation = {correlation:.3f})')
plt.xlabel('T1 (Biến độc lập)')
plt.ylabel('DH1 (Biến phụ thuộc)')
plt.grid(True, linestyle='--', alpha=0.5)

plt.show()
```

```python
#Hãy khảo sát tương quan giữa biến DH1 theo biến T1 trên từng nhóm khu vực
#  Tính hệ số tương quan theo từng nhóm KV
print("Hệ số tương quan T1 - DH1 theo từng khu vực:")
correlations = df.groupby('KV')[['T1', 'DH1']].corr().iloc[0::2, -1]
print(correlations)

#  Vẽ biểu đồ Scatter kèm đường hồi quy cho từng KV
# col='KV': Tách thành các biểu đồ con theo KV
# hue='KV': Tô màu khác nhau cho từng KV
sns.lmplot(x='T1', y='DH1', data=df, col='KV', hue='KV', col_wrap=3, height=4)

plt.subplots_adjust(top=0.85) # Chỉnh khoảng cách tiêu đề
plt.suptitle('Biểu đồ tương quan T1 và DH1 phân theo Khu vực')
plt.show()
```
